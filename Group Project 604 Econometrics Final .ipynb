### Model_1

```python
def model_summary(model):
    s = model.describe().T
    s['std_error'] = s['std']/np.sqrt(s['count'])
    s['tstat'] = s['mean']/s['std_error']
    return s[['mean','tstat']].iloc[:-2]
```

```python
model_1_summary = model_summary(model_1)
model_1_summary
```

```python
def add_asterisks(mean, t_statistic):
    if abs(t_statistic) >= 2.5834:
        return '{:.6f}{}'.format(mean, '***')
    elif abs(t_statistic) >= 1.9636:
        return '{:.6f}{}'.format(mean, '**')
    elif abs(t_statistic) >= 1.6472:
        return '{:.6f}{}'.format(mean, '*')
    else:
        return '{:.6f}'.format(mean)
```

```python
model_1_summary['mean'] = model_1_summary.apply(lambda row: add_asterisks(row['mean'], row['tstat']), axis=1)
model_1_summary_T = model_1_summary.T
model_1_summary_T
```

```python
def data_transform(model):
    M = pd.DataFrame(data = None)
    for i in model.columns:
        a = pd.DataFrame(model[i])
        a = a.rename(columns = {i:'Model_1'})
        a.index = ['Average of time series of ' + i,''] # set tstat as empty as required 
        M = M.append(a)
    return M
model_1_final = data_transform(model_1_summary_T)
model_1_final
```
